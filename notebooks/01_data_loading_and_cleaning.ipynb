```python
# Data Loading and Cleaning

## Purpose
This notebook loads the student performance dataset, checks its structure, identifies potential data quality issues, and prepares a cleaned version for later analysis and modeling.
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("../data/raw/student-mat.csv", sep=";")

# Display first 5 rows
df.head()

# Check shape
print("Shape of dataset:", df.shape)

# Check column names
print("\nColumns:")
print(df.columns.tolist())

# Check basic info
print("\nDataset info:")
df.info()

# Summary statistics for numerical columns
df.describe()

# Check missing values
missing_values = df.isnull().sum()
print("Missing values by column:")
print(missing_values)

print("\nTotal missing values in dataset:", df.isnull().sum().sum())

# Check duplicate rows
duplicate_count = df.duplicated().sum()
print("Number of duplicate rows:", duplicate_count)

# Inspect target variable
print("Summary of final grade (G3):")
print(df["G3"].describe())

# Check data types
print(df.dtypes)

# Create a copy for cleaning
cleaned_df = df.copy()

# Remove duplicates if any
cleaned_df = cleaned_df.drop_duplicates()

print("Shape after removing duplicates:", cleaned_df.shape)

# Save cleaned dataset
cleaned_df.to_csv("../data/processed/cleaned_student_mat.csv", index=False)

print("Cleaned dataset saved successfully.")

## Cleaning Summary
- The dataset was loaded successfully.
- Missing values were checked.
- Duplicate rows were checked and removed if necessary.
- A cleaned version of the dataset was saved for future analysis.
```
